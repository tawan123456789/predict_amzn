```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Regression Feature Importance
df_reg = pd.read_csv('AMZN_methods_all_in_one.csv')

# Sort data
df_reg_rf = df_reg.sort_values('rf_importance', ascending=True)
df_reg_kbest = df_reg.sort_values('selectkbest_fscore', ascending=True)

# Plot for Regression
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

axes[0].barh(df_reg_rf['feature'], df_reg_rf['rf_importance'], color='skyblue')
axes[0].set_xlabel('Random Forest Importance')
axes[0].set_title('RF Importance')
axes[0].grid(axis='x', alpha=0.3)

axes[1].barh(df_reg_kbest['feature'], df_reg_kbest['selectkbest_fscore'], color='lightgreen')
axes[1].set_xlabel('SelectKBest F-Score')
axes[1].set_title('SelectKBest Score')
axes[1].grid(axis='x', alpha=0.3)

plt.suptitle('Feature Importance for AMZN (Regression Method)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Classification Feature Importance
df_clf = pd.read_csv('CLASSIFY-AMZN_methods_all_in_one.csv')

# Sort data
df_clf_rf = df_clf.sort_values('rf_importance', ascending=True)
df_clf_kbest = df_clf.sort_values('selectkbest_fscore', ascending=True)

# Plot for Classification
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

axes[0].barh(df_clf_rf['feature'], df_clf_rf['rf_importance'], color='lightcoral')
axes[0].set_xlabel('Random Forest Importance')
axes[0].set_title('RF Importance')
axes[0].grid(axis='x', alpha=0.3)

axes[1].barh(df_clf_kbest['feature'], df_clf_kbest['selectkbest_fscore'], color='orange')
axes[1].set_xlabel('SelectKBest F-Score')
axes[1].set_title('SelectKBest Score')
axes[1].grid(axis='x', alpha=0.3)

plt.suptitle('Feature Importance for AMZN (Classification Method)')
plt.tight_layout()
plt.show()
```

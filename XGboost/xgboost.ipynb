```python
import os
import json
import math
import random
import logging
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBRegressor


# ============================================================
# 1) CONFIG
# ============================================================
CSV_PATH = "../SnP_daily_update_AMZN_features_with_target.csv"

DATE_COL = "Date"
TARGET_COL = "y_ret_t1"

TRAIN_RATIO = 0.70
VAL_RATIO   = 0.15
TEST_RATIO  = 0.15

assert abs(TRAIN_RATIO + VAL_RATIO + TEST_RATIO - 1.0) < 1e-9, "Ratios must sum to 1.0"

SEED = 42
RUN_TS = datetime.now().strftime("%Y%m%d_%H%M%S")

# ---------- Search config ----------
N_RANDOM_TRIALS = 160
SCREEN_TOP_K = 12
N_WF_SPLITS = 4
MIN_TRAIN_FRACTION = 0.45

# ---------- Early stopping ----------
EARLY_STOPPING_ROUNDS = 200
FINAL_ES_RATIO = 0.10

# ---------- Selection / filtering ----------
MIN_ACCEPT_STD_RATIO = 0.08
MAX_ACCEPT_STD_RATIO = 2.00

# ---------- Non-zero thresholds for collapse checks ----------
PRED_ABS_EPS_1 = 1e-4
PRED_ABS_EPS_2 = 5e-4
PRED_ABS_EPS_3 = 1e-3

# ---------- Output dirs ----------
OUT_DIR = f"xgb_amzn_improved_v2_{RUN_TS}"
LOG_DIR = os.path.join(OUT_DIR, "logs")
PLOT_DIR = os.path.join(OUT_DIR, "plots")
RESULT_DIR = os.path.join(OUT_DIR, "results")

for d in [OUT_DIR, LOG_DIR, PLOT_DIR, RESULT_DIR]:
    os.makedirs(d, exist_ok=True)

plt.rcParams["figure.figsize"] = (10, 5)


# ============================================================
# 2) LOGGER
# ============================================================
def build_logger(log_path: str, logger_name: str):
    logger = logging.getLogger(logger_name)
    logger.setLevel(logging.INFO)

    if logger.hasHandlers():
        logger.handlers.clear()

    fmt = logging.Formatter("%(asctime)s | %(levelname)s | %(message)s")

    fh = logging.FileHandler(log_path, encoding="utf-8")
    fh.setLevel(logging.INFO)
    fh.setFormatter(fmt)

    sh = logging.StreamHandler()
    sh.setLevel(logging.INFO)
    sh.setFormatter(fmt)

    logger.addHandler(fh)
    logger.addHandler(sh)
    return logger


main_log_path = os.path.join(LOG_DIR, "main.log")
logger = build_logger(main_log_path, "xgb_main_improved_v2")


# ============================================================
# 3) NOTEBOOK HELPER
# ============================================================
def save_and_show(fig, path=None):
    if path is not None:
        fig.savefig(path, dpi=150, bbox_inches="tight")
    plt.show()
    plt.close(fig)


# ============================================================
# 4) SEED
# ============================================================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)

set_seed(SEED)


# ============================================================
# 5) LOAD DATA
# ============================================================
df = pd.read_csv(CSV_PATH)
df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")
df = df.sort_values(DATE_COL).reset_index(drop=True)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
FEATURE_COLS = [c for c in numeric_cols if c != TARGET_COL]

use_cols = [DATE_COL] + FEATURE_COLS + [TARGET_COL]
df = df[use_cols].copy().dropna().reset_index(drop=True)

logger.info(f"Data shape after dropna: {df.shape}")
logger.info(f"Using {len(FEATURE_COLS)} features")
logger.info(f"Features: {FEATURE_COLS}")


# ============================================================
# 6) TIME SPLIT
# ============================================================
n = len(df)
train_end = int(n * TRAIN_RATIO)
val_end = int(n * (TRAIN_RATIO + VAL_RATIO))

train_df = df.iloc[:train_end].copy()
val_df   = df.iloc[train_end:val_end].copy()
test_df  = df.iloc[val_end:].copy()

train_val_df = pd.concat([train_df, val_df], axis=0).reset_index(drop=True)

logger.info(f"Train size:     {len(train_df)}")
logger.info(f"Val size:       {len(val_df)}")
logger.info(f"Train+Val size: {len(train_val_df)}")
logger.info(f"Test size:      {len(test_df)}")

logger.info(f"Train dates:     {train_df[DATE_COL].min()} -> {train_df[DATE_COL].max()}")
logger.info(f"Val dates:       {val_df[DATE_COL].min()} -> {val_df[DATE_COL].max()}")
logger.info(f"Train+Val dates: {train_val_df[DATE_COL].min()} -> {train_val_df[DATE_COL].max()}")
logger.info(f"Test dates:      {test_df[DATE_COL].min()} -> {test_df[DATE_COL].max()}")


# ============================================================
# 7) CLIP FEATURES BY TRAIN QUANTILES
# ============================================================
feature_clip_info = {}

for col in FEATURE_COLS:
    low = train_df[col].quantile(0.01)
    high = train_df[col].quantile(0.99)
    feature_clip_info[col] = {"low": float(low), "high": float(high)}

    train_df[col] = train_df[col].clip(low, high)
    val_df[col] = val_df[col].clip(low, high)
    train_val_df[col] = train_val_df[col].clip(low, high)
    test_df[col] = test_df[col].clip(low, high)

with open(os.path.join(RESULT_DIR, "feature_clip_info.json"), "w", encoding="utf-8") as f:
    json.dump(feature_clip_info, f, indent=2)


# ============================================================
# 8) PREPARE MATRICES
# ============================================================
X_train = train_df[FEATURE_COLS].copy()
y_train = train_df[TARGET_COL].copy()

X_val = val_df[FEATURE_COLS].copy()
y_val = val_df[TARGET_COL].copy()

X_train_val = train_val_df[FEATURE_COLS].copy()
y_train_val = train_val_df[TARGET_COL].copy()

X_test = test_df[FEATURE_COLS].copy()
y_test = test_df[TARGET_COL].copy()


# ============================================================
# 9) METRICS / SCORING
# ============================================================
def safe_corr(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    if len(y_true) < 2:
        return 0.0

    if np.std(y_true) < 1e-12 or np.std(y_pred) < 1e-12:
        return 0.0

    c = np.corrcoef(y_true, y_pred)[0, 1]
    if np.isnan(c):
        return 0.0
    return float(c)


def regression_metrics(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    direction_acc = float(np.mean(np.sign(y_true) == np.sign(y_pred)))

    pred_std = float(np.std(y_pred))
    true_std = float(np.std(y_true))
    std_ratio = float(pred_std / (true_std + 1e-12))
    corr = safe_corr(y_true, y_pred)

    pred_mean = float(np.mean(y_pred))
    true_mean = float(np.mean(y_true))
    pred_abs_mean = float(np.mean(np.abs(y_pred)))
    true_abs_mean = float(np.mean(np.abs(y_true)))
    abs_mean_ratio = float(pred_abs_mean / (true_abs_mean + 1e-12))

    pred_p05 = float(np.quantile(y_pred, 0.05))
    pred_p25 = float(np.quantile(y_pred, 0.25))
    pred_p50 = float(np.quantile(y_pred, 0.50))
    pred_p75 = float(np.quantile(y_pred, 0.75))
    pred_p95 = float(np.quantile(y_pred, 0.95))

    true_p05 = float(np.quantile(y_true, 0.05))
    true_p25 = float(np.quantile(y_true, 0.25))
    true_p50 = float(np.quantile(y_true, 0.50))
    true_p75 = float(np.quantile(y_true, 0.75))
    true_p95 = float(np.quantile(y_true, 0.95))

    nonzero_rate_1e4 = float(np.mean(np.abs(y_pred) > PRED_ABS_EPS_1))
    nonzero_rate_5e4 = float(np.mean(np.abs(y_pred) > PRED_ABS_EPS_2))
    nonzero_rate_1e3 = float(np.mean(np.abs(y_pred) > PRED_ABS_EPS_3))

    sign_pos_rate = float(np.mean(y_pred > 0))
    sign_neg_rate = float(np.mean(y_pred < 0))
    sign_zero_rate = float(np.mean(y_pred == 0))

    return {
        "rmse": rmse,
        "mae": mae,
        "r2": r2,
        "direction_acc": direction_acc,
        "pred_std": pred_std,
        "true_std": true_std,
        "std_ratio": std_ratio,
        "corr": corr,
        "pred_mean": pred_mean,
        "true_mean": true_mean,
        "pred_abs_mean": pred_abs_mean,
        "true_abs_mean": true_abs_mean,
        "abs_mean_ratio": abs_mean_ratio,
        "pred_p05": pred_p05,
        "pred_p25": pred_p25,
        "pred_p50": pred_p50,
        "pred_p75": pred_p75,
        "pred_p95": pred_p95,
        "true_p05": true_p05,
        "true_p25": true_p25,
        "true_p50": true_p50,
        "true_p75": true_p75,
        "true_p95": true_p95,
        "nonzero_rate_1e4": nonzero_rate_1e4,
        "nonzero_rate_5e4": nonzero_rate_5e4,
        "nonzero_rate_1e3": nonzero_rate_1e3,
        "sign_pos_rate": sign_pos_rate,
        "sign_neg_rate": sign_neg_rate,
        "sign_zero_rate": sign_zero_rate,
    }


def composite_score(metrics_dict):
    rmse = metrics_dict["rmse"]
    mae = metrics_dict["mae"]
    direction_acc = metrics_dict["direction_acc"]
    std_ratio = metrics_dict["std_ratio"]
    abs_mean_ratio = metrics_dict["abs_mean_ratio"]
    corr = metrics_dict["corr"]
    nonzero_rate_1e4 = metrics_dict["nonzero_rate_1e4"]
    nonzero_rate_5e4 = metrics_dict["nonzero_rate_5e4"]

    std_penalty = abs(1.0 - std_ratio)
    abs_mean_penalty = abs(1.0 - abs_mean_ratio)

    low_std_hard_penalty = 0.0
    if std_ratio < 0.20:
        low_std_hard_penalty += (0.20 - std_ratio) * 8.0
    if std_ratio < 0.10:
        low_std_hard_penalty += (0.10 - std_ratio) * 12.0
    if std_ratio < 0.05:
        low_std_hard_penalty += (0.05 - std_ratio) * 20.0

    high_std_penalty = 0.0
    if std_ratio > 1.80:
        high_std_penalty += (std_ratio - 1.80) * 2.0

    low_abs_mean_hard_penalty = 0.0
    if abs_mean_ratio < 0.20:
        low_abs_mean_hard_penalty += (0.20 - abs_mean_ratio) * 3.0
    if abs_mean_ratio < 0.10:
        low_abs_mean_hard_penalty += (0.10 - abs_mean_ratio) * 5.0

    collapse_penalty = 0.0
    if nonzero_rate_1e4 < 0.30:
        collapse_penalty += (0.30 - nonzero_rate_1e4) * 2.0
    if nonzero_rate_5e4 < 0.10:
        collapse_penalty += (0.10 - nonzero_rate_5e4) * 4.0

    score = (
        rmse
        + 0.25 * mae
        + 0.03 * std_penalty
        + 0.01 * abs_mean_penalty
        + low_std_hard_penalty
        + high_std_penalty
        + low_abs_mean_hard_penalty
        + collapse_penalty
        - 0.03 * direction_acc
        - 0.02 * corr
    )
    return float(score)


# ============================================================
# 9.1) VALIDATION PLOT HELPERS
# ============================================================
def plot_stage_validation_summary(df_results, out_path_prefix, top_n=15, stage_name="Stage"):
    if len(df_results) == 0:
        return {}

    top_df = df_results.head(top_n).copy()
    paths = {}

    fig = plt.figure(figsize=(12, 5))
    plt.bar(top_df["trial_name"], top_df["mean_score"])
    plt.xticks(rotation=75)
    plt.title(f"{stage_name}: Top {top_n} Trials by Mean Composite Score")
    plt.xlabel("Trial")
    plt.ylabel("Mean Composite Score")
    path = f"{out_path_prefix}_mean_score.png"
    save_and_show(fig, path)
    paths["mean_score_plot"] = path

    fig = plt.figure(figsize=(12, 5))
    plt.bar(top_df["trial_name"], top_df["mean_rmse"])
    plt.xticks(rotation=75)
    plt.title(f"{stage_name}: Top {top_n} Trials by Mean RMSE")
    plt.xlabel("Trial")
    plt.ylabel("Mean RMSE")
    path = f"{out_path_prefix}_mean_rmse.png"
    save_and_show(fig, path)
    paths["mean_rmse_plot"] = path

    fig = plt.figure(figsize=(12, 5))
    plt.bar(top_df["trial_name"], top_df["mean_direction_acc"])
    plt.xticks(rotation=75)
    plt.title(f"{stage_name}: Top {top_n} Trials by Mean Direction Accuracy")
    plt.xlabel("Trial")
    plt.ylabel("Mean Direction Accuracy")
    path = f"{out_path_prefix}_mean_direction_acc.png"
    save_and_show(fig, path)
    paths["mean_direction_acc_plot"] = path

    fig = plt.figure(figsize=(12, 5))
    plt.bar(top_df["trial_name"], top_df["mean_std_ratio"])
    plt.xticks(rotation=75)
    plt.axhline(MIN_ACCEPT_STD_RATIO, linestyle="--", label="Min Accept Std Ratio")
    plt.title(f"{stage_name}: Top {top_n} Trials by Mean Std Ratio")
    plt.xlabel("Trial")
    plt.ylabel("Mean Std Ratio")
    plt.legend()
    path = f"{out_path_prefix}_mean_std_ratio.png"
    save_and_show(fig, path)
    paths["mean_std_ratio_plot"] = path

    return paths


def plot_best_trial_folds(fold_df, out_path_prefix, trial_name="best_trial"):
    if len(fold_df) == 0:
        return {}

    fold_df = fold_df.copy().sort_values("fold")
    paths = {}

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["rmse"], marker="o")
    plt.title(f"{trial_name}: Validation RMSE by Fold")
    plt.xlabel("Fold")
    plt.ylabel("RMSE")
    plt.grid(True)
    path = f"{out_path_prefix}_fold_rmse.png"
    save_and_show(fig, path)
    paths["fold_rmse_plot"] = path

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["direction_acc"], marker="o")
    plt.axhline(0.5, linestyle="--")
    plt.title(f"{trial_name}: Validation Direction Accuracy by Fold")
    plt.xlabel("Fold")
    plt.ylabel("Direction Accuracy")
    plt.grid(True)
    path = f"{out_path_prefix}_fold_direction_acc.png"
    save_and_show(fig, path)
    paths["fold_direction_acc_plot"] = path

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["std_ratio"], marker="o")
    plt.axhline(MIN_ACCEPT_STD_RATIO, linestyle="--", label="Min Accept")
    plt.title(f"{trial_name}: Validation Std Ratio by Fold")
    plt.xlabel("Fold")
    plt.ylabel("Std Ratio")
    plt.legend()
    plt.grid(True)
    path = f"{out_path_prefix}_fold_std_ratio.png"
    save_and_show(fig, path)
    paths["fold_std_ratio_plot"] = path

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["score"], marker="o")
    plt.title(f"{trial_name}: Validation Composite Score by Fold")
    plt.xlabel("Fold")
    plt.ylabel("Score")
    plt.grid(True)
    path = f"{out_path_prefix}_fold_score.png"
    save_and_show(fig, path)
    paths["fold_score_plot"] = path

    return paths


def plot_final_learning_curve(evals_result, out_path):
    if not evals_result:
        return None

    metric_name = None
    if "validation_0" in evals_result and len(evals_result["validation_0"]) > 0:
        metric_name = list(evals_result["validation_0"].keys())[0]

    if metric_name is None:
        return None

    fig = plt.figure(figsize=(10, 5))

    if "validation_0" in evals_result:
        plt.plot(evals_result["validation_0"][metric_name], label="Train")
    if "validation_1" in evals_result:
        plt.plot(evals_result["validation_1"][metric_name], label="Eval")

    plt.title(f"Final Model Learning Curve ({metric_name})")
    plt.xlabel("Boosting Round")
    plt.ylabel(metric_name)
    plt.legend()
    plt.grid(True)

    save_and_show(fig, out_path)
    return out_path


# ============================================================
# 10) PARAM SPACE
# ============================================================
PARAM_SPACE = {
    "n_estimators": [500, 800, 1200, 1600, 2200, 3000],
    "max_depth": [2, 3, 4, 5, 6, 8, 10],
    "learning_rate": [0.005, 0.01, 0.02, 0.03, 0.05, 0.08],
    "subsample": [0.7, 0.85, 1.0],
    "colsample_bytree": [0.5, 0.7, 0.85, 1.0],
    "min_child_weight": [1, 2, 3, 5],
    "reg_alpha": [0.0, 0.0001, 0.001, 0.01, 0.1],
    "reg_lambda": [0.0, 0.1, 0.5, 1.0, 2.0, 5.0],
    "gamma": [0.0, 0.001, 0.01, 0.05, 0.1],
}


def sample_random_params(param_space, n_trials, seed=42):
    rng = random.Random(seed)
    seen = set()
    samples = []

    keys = list(param_space.keys())
    max_possible = 1
    for k in keys:
        max_possible *= len(param_space[k])

    n_trials = min(n_trials, max_possible)

    while len(samples) < n_trials:
        params = {k: rng.choice(param_space[k]) for k in keys}
        sig = tuple((k, params[k]) for k in keys)
        if sig not in seen:
            seen.add(sig)
            samples.append(params)

    return samples


random_param_list = sample_random_params(PARAM_SPACE, N_RANDOM_TRIALS, seed=SEED)
logger.info(f"Random search trials: {len(random_param_list)}")


# ============================================================
# 11) WALK-FORWARD SPLITS
# ============================================================
def make_expanding_window_splits(n_samples, n_splits=4, min_train_fraction=0.45):
    min_train_size = int(n_samples * min_train_fraction)
    remaining = n_samples - min_train_size
    fold_size = remaining // n_splits

    splits = []
    train_end = min_train_size

    for i in range(n_splits):
        if i == n_splits - 1:
            valid_end = n_samples
        else:
            valid_end = train_end + fold_size

        train_idx = np.arange(0, train_end)
        valid_idx = np.arange(train_end, valid_end)

        if len(valid_idx) > 0:
            splits.append((train_idx, valid_idx))

        train_end = valid_end

    return splits


wf_splits = make_expanding_window_splits(
    n_samples=len(train_val_df),
    n_splits=N_WF_SPLITS,
    min_train_fraction=MIN_TRAIN_FRACTION
)

logger.info(f"Number of walk-forward folds: {len(wf_splits)}")
for i, (tr_idx, va_idx) in enumerate(wf_splits, start=1):
    logger.info(
        f"Fold {i}: train={len(tr_idx)} rows ({train_val_df.iloc[tr_idx][DATE_COL].min()} -> {train_val_df.iloc[tr_idx][DATE_COL].max()}) | "
        f"valid={len(va_idx)} rows ({train_val_df.iloc[va_idx][DATE_COL].min()} -> {train_val_df.iloc[va_idx][DATE_COL].max()})"
    )


# ============================================================
# 12) BUILD MODEL
# ============================================================
def build_model(params):
    return XGBRegressor(
        n_estimators=int(params["n_estimators"]),
        max_depth=int(params["max_depth"]),
        learning_rate=float(params["learning_rate"]),
        subsample=float(params["subsample"]),
        colsample_bytree=float(params["colsample_bytree"]),
        min_child_weight=float(params["min_child_weight"]),
        reg_alpha=float(params["reg_alpha"]),
        reg_lambda=float(params["reg_lambda"]),
        gamma=float(params["gamma"]),
        objective="reg:squarederror",
        random_state=SEED,
        n_jobs=-1,
        early_stopping_rounds=EARLY_STOPPING_ROUNDS,
        tree_method="hist",
        eval_metric="rmse"
    )


# ============================================================
# 13) ONE WALK-FORWARD EVALUATION
# ============================================================
def evaluate_params_walk_forward(trial_name, params, X_all, y_all, date_series, splits):
    trial_log_path = os.path.join(LOG_DIR, f"{trial_name}.log")
    tlog = build_logger(trial_log_path, f"{trial_name}_logger")

    tlog.info("=" * 100)
    tlog.info(f"Start trial: {trial_name}")
    tlog.info(f"Params: {params}")

    fold_rows = []

    for fold_id, (tr_idx, va_idx) in enumerate(splits, start=1):
        X_tr = X_all.iloc[tr_idx]
        y_tr = y_all.iloc[tr_idx]
        X_va = X_all.iloc[va_idx]
        y_va = y_all.iloc[va_idx]

        model = build_model(params)
        model.fit(
            X_tr, y_tr,
            eval_set=[(X_va, y_va)],
            verbose=False
        )

        va_pred = model.predict(X_va)
        va_metrics = regression_metrics(y_va, va_pred)
        va_score = composite_score(va_metrics)

        best_iter = getattr(model, "best_iteration", None)
        if best_iter is None:
            best_iter = getattr(model, "best_ntree_limit", None)

        fold_row = {
            "trial_name": trial_name,
            "fold": fold_id,
            "train_start": str(date_series.iloc[tr_idx].min()),
            "train_end": str(date_series.iloc[tr_idx].max()),
            "valid_start": str(date_series.iloc[va_idx].min()),
            "valid_end": str(date_series.iloc[va_idx].max()),
            "best_iteration": best_iter,
            "rmse": va_metrics["rmse"],
            "mae": va_metrics["mae"],
            "r2": va_metrics["r2"],
            "direction_acc": va_metrics["direction_acc"],
            "pred_std": va_metrics["pred_std"],
            "true_std": va_metrics["true_std"],
            "std_ratio": va_metrics["std_ratio"],
            "corr": va_metrics["corr"],
            "pred_mean": va_metrics["pred_mean"],
            "true_mean": va_metrics["true_mean"],
            "pred_abs_mean": va_metrics["pred_abs_mean"],
            "true_abs_mean": va_metrics["true_abs_mean"],
            "abs_mean_ratio": va_metrics["abs_mean_ratio"],
            "nonzero_rate_1e4": va_metrics["nonzero_rate_1e4"],
            "nonzero_rate_5e4": va_metrics["nonzero_rate_5e4"],
            "nonzero_rate_1e3": va_metrics["nonzero_rate_1e3"],
            "score": va_score
        }
        fold_rows.append(fold_row)

        tlog.info(
            f"Fold {fold_id} | "
            f"rmse={va_metrics['rmse']:.6f} | "
            f"mae={va_metrics['mae']:.6f} | "
            f"r2={va_metrics['r2']:.6f} | "
            f"dir={va_metrics['direction_acc']:.4f} | "
            f"pred_std={va_metrics['pred_std']:.6f} | "
            f"true_std={va_metrics['true_std']:.6f} | "
            f"std_ratio={va_metrics['std_ratio']:.4f} | "
            f"abs_mean_ratio={va_metrics['abs_mean_ratio']:.4f} | "
            f"corr={va_metrics['corr']:.4f} | "
            f"nz@1e-4={va_metrics['nonzero_rate_1e4']:.4f} | "
            f"nz@5e-4={va_metrics['nonzero_rate_5e4']:.4f} | "
            f"nz@1e-3={va_metrics['nonzero_rate_1e3']:.4f} | "
            f"score={va_score:.6f} | "
            f"best_iteration={best_iter}"
        )

    fold_df = pd.DataFrame(fold_rows)

    agg = {
        "trial_name": trial_name,
        **params,
        "mean_rmse": fold_df["rmse"].mean(),
        "std_rmse": fold_df["rmse"].std(ddof=0),
        "mean_mae": fold_df["mae"].mean(),
        "mean_r2": fold_df["r2"].mean(),
        "mean_direction_acc": fold_df["direction_acc"].mean(),
        "mean_pred_std": fold_df["pred_std"].mean(),
        "mean_true_std": fold_df["true_std"].mean(),
        "mean_std_ratio": fold_df["std_ratio"].mean(),
        "mean_corr": fold_df["corr"].mean(),
        "mean_abs_mean_ratio": fold_df["abs_mean_ratio"].mean(),
        "mean_nonzero_rate_1e4": fold_df["nonzero_rate_1e4"].mean(),
        "mean_nonzero_rate_5e4": fold_df["nonzero_rate_5e4"].mean(),
        "mean_nonzero_rate_1e3": fold_df["nonzero_rate_1e3"].mean(),
        "mean_score": fold_df["score"].mean(),
        "log_path": trial_log_path
    }

    tlog.info(
        f"TRIAL SUMMARY | "
        f"mean_rmse={agg['mean_rmse']:.6f} | "
        f"mean_mae={agg['mean_mae']:.6f} | "
        f"mean_r2={agg['mean_r2']:.6f} | "
        f"mean_dir={agg['mean_direction_acc']:.4f} | "
        f"mean_std_ratio={agg['mean_std_ratio']:.4f} | "
        f"mean_abs_mean_ratio={agg['mean_abs_mean_ratio']:.4f} | "
        f"mean_corr={agg['mean_corr']:.4f} | "
        f"mean_nz@1e-4={agg['mean_nonzero_rate_1e4']:.4f} | "
        f"mean_nz@5e-4={agg['mean_nonzero_rate_5e4']:.4f} | "
        f"mean_nz@1e-3={agg['mean_nonzero_rate_1e3']:.4f} | "
        f"mean_score={agg['mean_score']:.6f}"
    )

    return agg, fold_df


# ============================================================
# 14) STAGE 1 SCREENING
# ============================================================
screen_results = []
all_fold_results = []

for i, params in enumerate(random_param_list, start=1):
    trial_name = f"trial_{i:03d}"

    logger.info("-" * 100)
    logger.info(f"Stage 1 screening: {trial_name}/{len(random_param_list)}")
    logger.info(f"Params: {params}")

    agg_result, fold_df = evaluate_params_walk_forward(
        trial_name=trial_name,
        params=params,
        X_all=X_train_val,
        y_all=y_train_val,
        date_series=train_val_df[DATE_COL],
        splits=wf_splits
    )

    screen_results.append(agg_result)
    all_fold_results.append(fold_df)

    screen_df_so_far = pd.DataFrame(screen_results).sort_values(
        ["mean_score", "mean_rmse", "mean_mae"],
        ascending=[True, True, True]
    )
    screen_df_so_far.to_csv(
        os.path.join(RESULT_DIR, "stage1_screening_results_so_far.csv"),
        index=False
    )

screen_df = pd.DataFrame(screen_results).sort_values(
    ["mean_score", "mean_rmse", "mean_mae"],
    ascending=[True, True, True]
).reset_index(drop=True)

screen_csv = os.path.join(RESULT_DIR, "stage1_screening_results.csv")
screen_df.to_csv(screen_csv, index=False)

fold_all_df = pd.concat(all_fold_results, axis=0).reset_index(drop=True)
fold_all_csv = os.path.join(RESULT_DIR, "stage1_screening_fold_results.csv")
fold_all_df.to_csv(fold_all_csv, index=False)

stage1_plot_paths = plot_stage_validation_summary(
    df_results=screen_df,
    out_path_prefix=os.path.join(PLOT_DIR, "stage1_validation"),
    top_n=15,
    stage_name="Stage 1"
)

stage_cols = [
    "trial_name",
    "mean_score",
    "mean_rmse",
    "mean_mae",
    "mean_r2",
    "mean_direction_acc",
    "mean_std_ratio",
    "mean_abs_mean_ratio",
    "mean_corr",
    "mean_nonzero_rate_1e4",
    "mean_nonzero_rate_5e4",
]

print("\n=== TOP 10 STAGE-1 TRIALS (RAW) ===")
display(screen_df[stage_cols].head(10))

screen_df_filtered = screen_df[
    (screen_df["mean_std_ratio"] >= MIN_ACCEPT_STD_RATIO) &
    (screen_df["mean_std_ratio"] <= MAX_ACCEPT_STD_RATIO)
].copy()

filtered_screen_csv = os.path.join(RESULT_DIR, "stage1_screening_results_filtered.csv")
screen_df_filtered.to_csv(filtered_screen_csv, index=False)

logger.info(f"Stage1 filtered candidates: {len(screen_df_filtered)} / {len(screen_df)}")

if len(screen_df_filtered) == 0:
    logger.info("No candidate passed std-ratio filter. Fallback to raw ranking.")
    top_k_df = screen_df.head(SCREEN_TOP_K).copy()
else:
    top_k_df = screen_df_filtered.head(SCREEN_TOP_K).copy()

print("\n=== TOP STAGE-1 TRIALS AFTER FILTER ===")
display(top_k_df[stage_cols].head(10))


# ============================================================
# 15) STAGE 2: RECHECK TOP-K
# ============================================================
stage2_results = []
stage2_fold_results = []

for _, row in top_k_df.iterrows():
    params = {
        "n_estimators": int(row["n_estimators"]),
        "max_depth": int(row["max_depth"]),
        "learning_rate": float(row["learning_rate"]),
        "subsample": float(row["subsample"]),
        "colsample_bytree": float(row["colsample_bytree"]),
        "min_child_weight": float(row["min_child_weight"]),
        "reg_alpha": float(row["reg_alpha"]),
        "reg_lambda": float(row["reg_lambda"]),
        "gamma": float(row["gamma"]),
    }

    trial_name = f"stage2_{row['trial_name']}"
    logger.info(f"Stage 2 recheck: {trial_name}")

    agg_result, fold_df = evaluate_params_walk_forward(
        trial_name=trial_name,
        params=params,
        X_all=X_train_val,
        y_all=y_train_val,
        date_series=train_val_df[DATE_COL],
        splits=wf_splits
    )

    stage2_results.append(agg_result)
    stage2_fold_results.append(fold_df)

stage2_df = pd.DataFrame(stage2_results).sort_values(
    ["mean_score", "mean_rmse", "mean_mae"],
    ascending=[True, True, True]
).reset_index(drop=True)

stage2_csv = os.path.join(RESULT_DIR, "stage2_topk_results.csv")
stage2_df.to_csv(stage2_csv, index=False)

stage2_fold_df = pd.concat(stage2_fold_results, axis=0).reset_index(drop=True)
stage2_fold_csv = os.path.join(RESULT_DIR, "stage2_topk_fold_results.csv")
stage2_fold_df.to_csv(stage2_fold_csv, index=False)

stage2_plot_paths = plot_stage_validation_summary(
    df_results=stage2_df,
    out_path_prefix=os.path.join(PLOT_DIR, "stage2_validation"),
    top_n=min(12, len(stage2_df)),
    stage_name="Stage 2"
)

print("\n=== STAGE-2 RESULTS ===")
display(stage2_df[stage_cols].head(12))

stage2_df_filtered = stage2_df[
    (stage2_df["mean_std_ratio"] >= MIN_ACCEPT_STD_RATIO) &
    (stage2_df["mean_std_ratio"] <= MAX_ACCEPT_STD_RATIO)
].copy()

stage2_filtered_csv = os.path.join(RESULT_DIR, "stage2_topk_results_filtered.csv")
stage2_df_filtered.to_csv(stage2_filtered_csv, index=False)

logger.info(f"Stage2 filtered candidates: {len(stage2_df_filtered)} / {len(stage2_df)}")

if len(stage2_df_filtered) > 0:
    best_row = stage2_df_filtered.sort_values(
        ["mean_score", "mean_rmse", "mean_mae"],
        ascending=[True, True, True]
    ).iloc[0].copy()
    logger.info("Best row selected from filtered stage2 results.")
else:
    best_row = stage2_df.sort_values(
        ["mean_score", "mean_rmse", "mean_mae"],
        ascending=[True, True, True]
    ).iloc[0].copy()
    logger.info("No filtered stage2 candidate. Best row selected from raw stage2 results.")

best_params = {
    "n_estimators": int(best_row["n_estimators"]),
    "max_depth": int(best_row["max_depth"]),
    "learning_rate": float(best_row["learning_rate"]),
    "subsample": float(best_row["subsample"]),
    "colsample_bytree": float(best_row["colsample_bytree"]),
    "min_child_weight": float(best_row["min_child_weight"]),
    "reg_alpha": float(best_row["reg_alpha"]),
    "reg_lambda": float(best_row["reg_lambda"]),
    "gamma": float(best_row["gamma"]),
}

logger.info("=" * 100)
logger.info(f"Best params after stage 2: {best_params}")
logger.info(f"Best stage2 trial: {best_row['trial_name']}")

best_stage2_fold_df = stage2_fold_df[
    stage2_fold_df["trial_name"] == best_row["trial_name"]
].copy()

best_trial_plot_paths = plot_best_trial_folds(
    fold_df=best_stage2_fold_df,
    out_path_prefix=os.path.join(PLOT_DIR, f"{best_row['trial_name']}_validation"),
    trial_name=best_row["trial_name"]
)

print("\n=== BEST TRIAL SUMMARY ===")
display(pd.DataFrame([best_row]))


# ============================================================
# 16) FINAL TRAIN ON TRAIN+VAL
# ============================================================
def fit_final_model(params, X_all, y_all, date_all, final_es_ratio=0.10):
    n_all = len(X_all)
    es_size = max(50, int(n_all * final_es_ratio))
    train_size = n_all - es_size

    X_fit = X_all.iloc[:train_size]
    y_fit = y_all.iloc[:train_size]
    X_es = X_all.iloc[train_size:]
    y_es = y_all.iloc[train_size:]

    logger.info(f"Final fit train size: {len(X_fit)}")
    logger.info(f"Final fit eval size : {len(X_es)}")
    logger.info(f"Final fit train dates: {date_all.iloc[:train_size].min()} -> {date_all.iloc[:train_size].max()}")
    logger.info(f"Final fit eval dates : {date_all.iloc[train_size:].min()} -> {date_all.iloc[train_size:].max()}")

    model = build_model(params)
    model.fit(
        X_fit, y_fit,
        eval_set=[(X_fit, y_fit), (X_es, y_es)],
        verbose=False
    )

    evals_result = model.evals_result()
    return model, train_size, es_size, evals_result


best_model, final_train_size, final_es_size, final_evals_result = fit_final_model(
    params=best_params,
    X_all=X_train_val,
    y_all=y_train_val,
    date_all=train_val_df[DATE_COL],
    final_es_ratio=FINAL_ES_RATIO
)

best_iteration = getattr(best_model, "best_iteration", None)
logger.info(f"Final model best_iteration: {best_iteration}")

final_learning_curve_path = plot_final_learning_curve(
    evals_result=final_evals_result,
    out_path=os.path.join(PLOT_DIR, "final_learning_curve.png")
)


# ============================================================
# 17) TEST EVALUATION
# ============================================================
test_pred = best_model.predict(X_test)
test_metrics = regression_metrics(y_test, test_pred)
test_score = composite_score(test_metrics)

print("\n=== FINAL TEST RESULT ===")
print(f"Best trial          : {best_row['trial_name']}")
print(f"Test RMSE           : {test_metrics['rmse']:.6f}")
print(f"Test MAE            : {test_metrics['mae']:.6f}")
print(f"Test R2             : {test_metrics['r2']:.6f}")
print(f"Test Direction Acc  : {test_metrics['direction_acc']:.4f}")
print(f"Pred std            : {test_metrics['pred_std']:.6f}")
print(f"True std            : {test_metrics['true_std']:.6f}")
print(f"Std ratio           : {test_metrics['std_ratio']:.4f}")
print(f"Abs mean ratio      : {test_metrics['abs_mean_ratio']:.4f}")
print(f"Corr                : {test_metrics['corr']:.4f}")
print(f"Nonzero rate >1e-4  : {test_metrics['nonzero_rate_1e4']:.4f}")
print(f"Nonzero rate >5e-4  : {test_metrics['nonzero_rate_5e4']:.4f}")
print(f"Nonzero rate >1e-3  : {test_metrics['nonzero_rate_1e3']:.4f}")
print(f"Composite score     : {test_score:.6f}")

logger.info("=" * 100)
logger.info("FINAL TEST RESULT")
logger.info(f"Best trial          : {best_row['trial_name']}")
logger.info(f"Test RMSE           : {test_metrics['rmse']:.6f}")
logger.info(f"Test MAE            : {test_metrics['mae']:.6f}")
logger.info(f"Test R2             : {test_metrics['r2']:.6f}")
logger.info(f"Test Direction Acc  : {test_metrics['direction_acc']:.4f}")
logger.info(f"Test Pred std       : {test_metrics['pred_std']:.6f}")
logger.info(f"Test True std       : {test_metrics['true_std']:.6f}")
logger.info(f"Test Std ratio      : {test_metrics['std_ratio']:.4f}")
logger.info(f"Test AbsMean ratio  : {test_metrics['abs_mean_ratio']:.4f}")
logger.info(f"Test Corr           : {test_metrics['corr']:.4f}")
logger.info(f"Test Nonzero >1e-4  : {test_metrics['nonzero_rate_1e4']:.4f}")
logger.info(f"Test Nonzero >5e-4  : {test_metrics['nonzero_rate_5e4']:.4f}")
logger.info(f"Test Nonzero >1e-3  : {test_metrics['nonzero_rate_1e3']:.4f}")
logger.info(f"Test Composite      : {test_score:.6f}")


# ============================================================
# 18) SAVE TEST PREDICTIONS
# ============================================================
pred_df = pd.DataFrame({
    DATE_COL: test_df[DATE_COL].values,
    "Actual_y_ret_t1": y_test.values,
    "Predicted_y_ret_t1": test_pred,
})
pred_df["Residual"] = pred_df["Actual_y_ret_t1"] - pred_df["Predicted_y_ret_t1"]
pred_df["ActualSign"] = np.sign(pred_df["Actual_y_ret_t1"])
pred_df["PredSign"] = np.sign(pred_df["Predicted_y_ret_t1"])
pred_df["Hit"] = (pred_df["ActualSign"] == pred_df["PredSign"]).astype(int)

pred_csv = os.path.join(RESULT_DIR, "test_predictions.csv")
pred_df.to_csv(pred_csv, index=False)

print("\n=== TEST PREDICTIONS (HEAD) ===")
display(pred_df.head(20))


# ============================================================
# 19) FEATURE IMPORTANCE
# ============================================================
importance_df = pd.DataFrame({
    "feature": FEATURE_COLS,
    "importance": best_model.feature_importances_
}).sort_values("importance", ascending=False).reset_index(drop=True)

importance_csv = os.path.join(RESULT_DIR, "feature_importance.csv")
importance_df.to_csv(importance_csv, index=False)

print("\n=== TOP 20 FEATURE IMPORTANCE ===")
display(importance_df.head(20))

fig = plt.figure(figsize=(10, 8))
top_imp = importance_df.head(20).iloc[::-1]
plt.barh(top_imp["feature"], top_imp["importance"])
plt.title("Top 20 Feature Importances")
plt.xlabel("Importance")
save_and_show(fig, os.path.join(PLOT_DIR, "feature_importance_top20.png"))


# ============================================================
# 20) PLOTS
# ============================================================
fig = plt.figure(figsize=(12, 5))
plt.plot(pred_df["Actual_y_ret_t1"].values[:200], label="Actual")
plt.plot(pred_df["Predicted_y_ret_t1"].values[:200], label="Predicted")
plt.title("Actual vs Predicted y_ret_t1 (first 200 test samples)")
plt.xlabel("Sample")
plt.ylabel("y_ret_t1")
plt.legend()
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "test_actual_vs_pred.png"))

fig = plt.figure(figsize=(6, 6))
plt.scatter(pred_df["Actual_y_ret_t1"], pred_df["Predicted_y_ret_t1"], alpha=0.5)
mn = min(pred_df["Actual_y_ret_t1"].min(), pred_df["Predicted_y_ret_t1"].min())
mx = max(pred_df["Actual_y_ret_t1"].max(), pred_df["Predicted_y_ret_t1"].max())
plt.plot([mn, mx], [mn, mx], linestyle="--")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Scatter: Actual vs Predicted")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "scatter_actual_vs_pred.png"))

fig = plt.figure(figsize=(12, 4))
plt.plot(pred_df["Residual"].values)
plt.axhline(0, linestyle="--")
plt.title("Residuals over Test Samples")
plt.xlabel("Sample")
plt.ylabel("Residual")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "residual_over_time.png"))

fig = plt.figure(figsize=(8, 4))
plt.hist(pred_df["Residual"].values, bins=40)
plt.title("Residual Histogram")
plt.xlabel("Residual")
plt.ylabel("Count")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "residual_hist.png"))

fig = plt.figure(figsize=(8, 4))
plt.hist(pred_df["Actual_y_ret_t1"].values, bins=40, alpha=0.6, label="Actual")
plt.hist(pred_df["Predicted_y_ret_t1"].values, bins=40, alpha=0.6, label="Predicted")
plt.title("Actual vs Predicted Distribution")
plt.xlabel("Return")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "actual_vs_pred_distribution.png"))

pred_df["CumHitRate"] = pred_df["Hit"].expanding().mean()
fig = plt.figure(figsize=(12, 4))
plt.plot(pred_df["CumHitRate"].values)
plt.axhline(0.5, linestyle="--")
plt.title("Cumulative Direction Accuracy")
plt.xlabel("Sample")
plt.ylabel("Accuracy")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "cumulative_direction_accuracy.png"))

pred_df["RollingPredStd_30"] = pred_df["Predicted_y_ret_t1"].rolling(30).std()
pred_df["RollingTrueStd_30"] = pred_df["Actual_y_ret_t1"].rolling(30).std()
fig = plt.figure(figsize=(12, 4))
plt.plot(pred_df["RollingPredStd_30"].values, label="Pred rolling std (30)")
plt.plot(pred_df["RollingTrueStd_30"].values, label="True rolling std (30)")
plt.title("Rolling Std Comparison")
plt.xlabel("Sample")
plt.ylabel("Std")
plt.legend()
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "rolling_std_comparison.png"))


# ============================================================
# 21) SAVE FINAL SUMMARY
# ============================================================
final_summary = {
    "num_features": len(FEATURE_COLS),
    "feature_cols": FEATURE_COLS,
    "search_trials": N_RANDOM_TRIALS,
    "screen_top_k": SCREEN_TOP_K,
    "n_walk_forward_splits": N_WF_SPLITS,
    "min_accept_std_ratio": MIN_ACCEPT_STD_RATIO,
    "max_accept_std_ratio": MAX_ACCEPT_STD_RATIO,
    "best_trial_name": str(best_row["trial_name"]),
    "best_params": best_params,
    "best_stage2_mean_score": float(best_row["mean_score"]),
    "best_stage2_mean_rmse": float(best_row["mean_rmse"]),
    "best_stage2_mean_mae": float(best_row["mean_mae"]),
    "best_stage2_mean_r2": float(best_row["mean_r2"]),
    "best_stage2_mean_direction_acc": float(best_row["mean_direction_acc"]),
    "best_stage2_mean_std_ratio": float(best_row["mean_std_ratio"]),
    "best_stage2_mean_abs_mean_ratio": float(best_row["mean_abs_mean_ratio"]),
    "best_stage2_mean_corr": float(best_row["mean_corr"]),
    "best_stage2_mean_nonzero_rate_1e4": float(best_row["mean_nonzero_rate_1e4"]),
    "best_stage2_mean_nonzero_rate_5e4": float(best_row["mean_nonzero_rate_5e4"]),
    "best_stage2_mean_nonzero_rate_1e3": float(best_row["mean_nonzero_rate_1e3"]),
    "final_best_iteration": None if best_iteration is None else int(best_iteration),
    "final_train_size_before_es": int(final_train_size),
    "final_es_size": int(final_es_size),
    "test_rmse": float(test_metrics["rmse"]),
    "test_mae": float(test_metrics["mae"]),
    "test_r2": float(test_metrics["r2"]),
    "test_direction_acc": float(test_metrics["direction_acc"]),
    "test_pred_std": float(test_metrics["pred_std"]),
    "test_true_std": float(test_metrics["true_std"]),
    "test_std_ratio": float(test_metrics["std_ratio"]),
    "test_abs_mean_ratio": float(test_metrics["abs_mean_ratio"]),
    "test_corr": float(test_metrics["corr"]),
    "test_nonzero_rate_1e4": float(test_metrics["nonzero_rate_1e4"]),
    "test_nonzero_rate_5e4": float(test_metrics["nonzero_rate_5e4"]),
    "test_nonzero_rate_1e3": float(test_metrics["nonzero_rate_1e3"]),
    "test_composite_score": float(test_score),
    "stage1_screening_csv": screen_csv,
    "stage1_screening_filtered_csv": filtered_screen_csv,
    "stage1_fold_csv": fold_all_csv,
    "stage2_topk_csv": stage2_csv,
    "stage2_topk_filtered_csv": stage2_filtered_csv,
    "stage2_fold_csv": stage2_fold_csv,
    "pred_csv": pred_csv,
    "feature_importance_csv": importance_csv,
    "main_log_path": main_log_path,
    "stage1_validation_plots": stage1_plot_paths,
    "stage2_validation_plots": stage2_plot_paths,
    "best_trial_validation_plots": best_trial_plot_paths,
    "final_learning_curve_plot": final_learning_curve_path
}

with open(os.path.join(RESULT_DIR, "final_summary.json"), "w", encoding="utf-8") as f:
    json.dump(final_summary, f, indent=2)

print("\n=== FINAL SUMMARY ===")
display(pd.DataFrame([final_summary]))

logger.info("All done.")
```

```python
import os
import json
import math
import random
import logging
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display
from xgboost import XGBRegressor


# ============================================================
# 1) CONFIG
# ============================================================
CSV_PATH = "../SnP_daily_update_AMZN_features_with_target.csv"

DATE_COL = "Date"
TARGET_COL = "y_ret_t1"

TRAIN_RATIO = 0.70
VAL_RATIO   = 0.15
TEST_RATIO  = 0.15

assert abs(TRAIN_RATIO + VAL_RATIO + TEST_RATIO - 1.0) < 1e-9, "Ratios must sum to 1.0"

SEED = 42
RUN_TS = datetime.now().strftime("%Y%m%d_%H%M%S")

# ---------- Walk-forward ----------
N_WF_SPLITS = 4
MIN_TRAIN_FRACTION = 0.45

# ---------- Early stopping ----------
EARLY_STOPPING_ROUNDS = 200
FINAL_ES_RATIO = 0.10

# ---------- Selection / filtering ----------
MIN_ACCEPT_STD_RATIO = 0.08
MAX_ACCEPT_STD_RATIO = 2.00

# ---------- Non-zero thresholds ----------
PRED_ABS_EPS_1 = 1e-4
PRED_ABS_EPS_2 = 5e-4
PRED_ABS_EPS_3 = 1e-3

# ---------- Fixed single-trial params ----------
BEST_PARAMS = {
    "n_estimators": 1600,
    "max_depth": 10,
    "learning_rate": 0.05,
    "subsample": 0.85,
    "colsample_bytree": 0.7,
    "min_child_weight": 5.0,
    "reg_alpha": 0.01,
    "reg_lambda": 0.5,
    "gamma": 0.001
}

# ---------- Output dirs ----------
OUT_DIR = f"xgb_amzn_single_trial_{RUN_TS}"
LOG_DIR = os.path.join(OUT_DIR, "logs")
PLOT_DIR = os.path.join(OUT_DIR, "plots")
RESULT_DIR = os.path.join(OUT_DIR, "results")

for d in [OUT_DIR, LOG_DIR, PLOT_DIR, RESULT_DIR]:
    os.makedirs(d, exist_ok=True)

plt.rcParams["figure.figsize"] = (10, 5)


# ============================================================
# 2) LOGGER
# ============================================================
def build_logger(log_path: str, logger_name: str):
    logger = logging.getLogger(logger_name)
    logger.setLevel(logging.INFO)

    if logger.hasHandlers():
        logger.handlers.clear()

    fmt = logging.Formatter("%(asctime)s | %(levelname)s | %(message)s")

    fh = logging.FileHandler(log_path, encoding="utf-8")
    fh.setLevel(logging.INFO)
    fh.setFormatter(fmt)

    sh = logging.StreamHandler()
    sh.setLevel(logging.INFO)
    sh.setFormatter(fmt)

    logger.addHandler(fh)
    logger.addHandler(sh)
    return logger


main_log_path = os.path.join(LOG_DIR, "main.log")
logger = build_logger(main_log_path, "xgb_single_trial_main")


# ============================================================
# 3) NOTEBOOK HELPER
# ============================================================
def save_and_show(fig, path=None):
    if path is not None:
        fig.savefig(path, dpi=150, bbox_inches="tight")
    plt.show()
    plt.close(fig)


# ============================================================
# 4) SEED
# ============================================================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)

set_seed(SEED)


# ============================================================
# 5) METRICS / SCORING (FROM SCRATCH)
# ============================================================
def mse_from_scratch(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    diff = y_true - y_pred
    return float(np.mean(diff * diff))


def rmse_from_scratch(y_true, y_pred):
    return float(math.sqrt(mse_from_scratch(y_true, y_pred)))


def mae_from_scratch(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.mean(np.abs(y_true - y_pred)))


def r2_from_scratch(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)

    ss_res = float(np.sum((y_true - y_pred) ** 2))
    y_mean = float(np.mean(y_true))
    ss_tot = float(np.sum((y_true - y_mean) ** 2))

    if ss_tot < 1e-12:
        return 0.0

    return float(1.0 - (ss_res / ss_tot))


def safe_corr(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)

    if len(y_true) < 2:
        return 0.0

    y_true_mean = float(np.mean(y_true))
    y_pred_mean = float(np.mean(y_pred))

    y_true_centered = y_true - y_true_mean
    y_pred_centered = y_pred - y_pred_mean

    num = float(np.sum(y_true_centered * y_pred_centered))
    den_left = float(np.sum(y_true_centered ** 2))
    den_right = float(np.sum(y_pred_centered ** 2))
    den = math.sqrt(den_left * den_right)

    if den < 1e-12:
        return 0.0

    return float(num / den)


def regression_metrics(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)

    rmse = rmse_from_scratch(y_true, y_pred)
    mae = mae_from_scratch(y_true, y_pred)
    r2 = r2_from_scratch(y_true, y_pred)

    direction_acc = float(np.mean(np.sign(y_true) == np.sign(y_pred)))

    pred_std = float(np.std(y_pred))
    true_std = float(np.std(y_true))
    std_ratio = float(pred_std / (true_std + 1e-12))
    corr = safe_corr(y_true, y_pred)

    pred_mean = float(np.mean(y_pred))
    true_mean = float(np.mean(y_true))
    pred_abs_mean = float(np.mean(np.abs(y_pred)))
    true_abs_mean = float(np.mean(np.abs(y_true)))
    abs_mean_ratio = float(pred_abs_mean / (true_abs_mean + 1e-12))

    nonzero_rate_1e4 = float(np.mean(np.abs(y_pred) > PRED_ABS_EPS_1))
    nonzero_rate_5e4 = float(np.mean(np.abs(y_pred) > PRED_ABS_EPS_2))
    nonzero_rate_1e3 = float(np.mean(np.abs(y_pred) > PRED_ABS_EPS_3))

    return {
        "rmse": rmse,
        "mae": mae,
        "r2": r2,
        "direction_acc": direction_acc,
        "pred_std": pred_std,
        "true_std": true_std,
        "std_ratio": std_ratio,
        "corr": corr,
        "pred_mean": pred_mean,
        "true_mean": true_mean,
        "pred_abs_mean": pred_abs_mean,
        "true_abs_mean": true_abs_mean,
        "abs_mean_ratio": abs_mean_ratio,
        "nonzero_rate_1e4": nonzero_rate_1e4,
        "nonzero_rate_5e4": nonzero_rate_5e4,
        "nonzero_rate_1e3": nonzero_rate_1e3,
    }


def composite_score(metrics_dict):
    rmse = metrics_dict["rmse"]
    mae = metrics_dict["mae"]
    direction_acc = metrics_dict["direction_acc"]
    std_ratio = metrics_dict["std_ratio"]
    abs_mean_ratio = metrics_dict["abs_mean_ratio"]
    corr = metrics_dict["corr"]
    nonzero_rate_1e4 = metrics_dict["nonzero_rate_1e4"]
    nonzero_rate_5e4 = metrics_dict["nonzero_rate_5e4"]

    std_penalty = abs(1.0 - std_ratio)
    abs_mean_penalty = abs(1.0 - abs_mean_ratio)

    low_std_hard_penalty = 0.0
    if std_ratio < 0.20:
        low_std_hard_penalty += (0.20 - std_ratio) * 8.0
    if std_ratio < 0.10:
        low_std_hard_penalty += (0.10 - std_ratio) * 12.0
    if std_ratio < 0.05:
        low_std_hard_penalty += (0.05 - std_ratio) * 20.0

    high_std_penalty = 0.0
    if std_ratio > 1.80:
        high_std_penalty += (std_ratio - 1.80) * 2.0

    low_abs_mean_hard_penalty = 0.0
    if abs_mean_ratio < 0.20:
        low_abs_mean_hard_penalty += (0.20 - abs_mean_ratio) * 3.0
    if abs_mean_ratio < 0.10:
        low_abs_mean_hard_penalty += (0.10 - abs_mean_ratio) * 5.0

    collapse_penalty = 0.0
    if nonzero_rate_1e4 < 0.30:
        collapse_penalty += (0.30 - nonzero_rate_1e4) * 2.0
    if nonzero_rate_5e4 < 0.10:
        collapse_penalty += (0.10 - nonzero_rate_5e4) * 4.0

    score = (
        rmse
        + 0.25 * mae
        + 0.03 * std_penalty
        + 0.01 * abs_mean_penalty
        + low_std_hard_penalty
        + high_std_penalty
        + low_abs_mean_hard_penalty
        + collapse_penalty
        - 0.03 * direction_acc
        - 0.02 * corr
    )
    return float(score)


# ============================================================
# 6) LOAD DATA
# ============================================================
df = pd.read_csv(CSV_PATH)
df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")
df = df.sort_values(DATE_COL).reset_index(drop=True)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
FEATURE_COLS = [c for c in numeric_cols if c != TARGET_COL]

use_cols = [DATE_COL] + FEATURE_COLS + [TARGET_COL]
df = df[use_cols].copy().dropna().reset_index(drop=True)

logger.info(f"Data shape after dropna: {df.shape}")
logger.info(f"Using {len(FEATURE_COLS)} features")
logger.info(f"Features: {FEATURE_COLS}")


# ============================================================
# 7) TIME SPLIT
# ============================================================
n = len(df)
train_end = int(n * TRAIN_RATIO)
val_end = int(n * (TRAIN_RATIO + VAL_RATIO))

train_df = df.iloc[:train_end].copy()
val_df   = df.iloc[train_end:val_end].copy()
test_df  = df.iloc[val_end:].copy()

train_val_df = pd.concat([train_df, val_df], axis=0).reset_index(drop=True)

logger.info(f"Train size:     {len(train_df)}")
logger.info(f"Val size:       {len(val_df)}")
logger.info(f"Train+Val size: {len(train_val_df)}")
logger.info(f"Test size:      {len(test_df)}")


# ============================================================
# 8) CLIP FEATURES BY TRAIN QUANTILES
# ============================================================
feature_clip_info = {}

for col in FEATURE_COLS:
    low = train_df[col].quantile(0.01)
    high = train_df[col].quantile(0.99)
    feature_clip_info[col] = {"low": float(low), "high": float(high)}

    train_df[col] = train_df[col].clip(low, high)
    val_df[col] = val_df[col].clip(low, high)
    train_val_df[col] = train_val_df[col].clip(low, high)
    test_df[col] = test_df[col].clip(low, high)

with open(os.path.join(RESULT_DIR, "feature_clip_info.json"), "w", encoding="utf-8") as f:
    json.dump(feature_clip_info, f, indent=2)


# ============================================================
# 9) PREPARE MATRICES
# ============================================================
X_train = train_df[FEATURE_COLS].copy()
y_train = train_df[TARGET_COL].copy()

X_val = val_df[FEATURE_COLS].copy()
y_val = val_df[TARGET_COL].copy()

X_train_val = train_val_df[FEATURE_COLS].copy()
y_train_val = train_val_df[TARGET_COL].copy()

X_test = test_df[FEATURE_COLS].copy()
y_test = test_df[TARGET_COL].copy()


# ============================================================
# 10) VALIDATION PLOT HELPERS
# ============================================================
def plot_best_trial_folds(fold_df, out_path_prefix, trial_name="best_trial"):
    if len(fold_df) == 0:
        return {}

    fold_df = fold_df.copy().sort_values("fold")
    paths = {}

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["rmse"], marker="o")
    plt.title(f"{trial_name}: Validation RMSE by Fold")
    plt.xlabel("Fold")
    plt.ylabel("RMSE")
    plt.grid(True)
    path = f"{out_path_prefix}_fold_rmse.png"
    save_and_show(fig, path)
    paths["fold_rmse_plot"] = path

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["direction_acc"], marker="o")
    plt.axhline(0.5, linestyle="--")
    plt.title(f"{trial_name}: Validation Direction Accuracy by Fold")
    plt.xlabel("Fold")
    plt.ylabel("Direction Accuracy")
    plt.grid(True)
    path = f"{out_path_prefix}_fold_direction_acc.png"
    save_and_show(fig, path)
    paths["fold_direction_acc_plot"] = path

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["std_ratio"], marker="o")
    plt.axhline(MIN_ACCEPT_STD_RATIO, linestyle="--", label="Min Accept")
    plt.title(f"{trial_name}: Validation Std Ratio by Fold")
    plt.xlabel("Fold")
    plt.ylabel("Std Ratio")
    plt.legend()
    plt.grid(True)
    path = f"{out_path_prefix}_fold_std_ratio.png"
    save_and_show(fig, path)
    paths["fold_std_ratio_plot"] = path

    fig = plt.figure(figsize=(10, 4))
    plt.plot(fold_df["fold"], fold_df["score"], marker="o")
    plt.title(f"{trial_name}: Validation Composite Score by Fold")
    plt.xlabel("Fold")
    plt.ylabel("Score")
    plt.grid(True)
    path = f"{out_path_prefix}_fold_score.png"
    save_and_show(fig, path)
    paths["fold_score_plot"] = path

    return paths


def plot_final_learning_curve(evals_result, out_path):
    if not evals_result:
        return None

    metric_name = None
    if "validation_0" in evals_result and len(evals_result["validation_0"]) > 0:
        metric_name = list(evals_result["validation_0"].keys())[0]

    if metric_name is None:
        return None

    fig = plt.figure(figsize=(10, 5))

    if "validation_0" in evals_result:
        plt.plot(evals_result["validation_0"][metric_name], label="Train")
    if "validation_1" in evals_result:
        plt.plot(evals_result["validation_1"][metric_name], label="Eval")

    plt.title(f"Final Model Learning Curve ({metric_name})")
    plt.xlabel("Boosting Round")
    plt.ylabel(metric_name)
    plt.legend()
    plt.grid(True)

    save_and_show(fig, out_path)
    return out_path


# ============================================================
# 11) WALK-FORWARD SPLITS
# ============================================================
def make_expanding_window_splits(n_samples, n_splits=4, min_train_fraction=0.45):
    min_train_size = int(n_samples * min_train_fraction)
    remaining = n_samples - min_train_size
    fold_size = remaining // n_splits

    splits = []
    train_end = min_train_size

    for i in range(n_splits):
        if i == n_splits - 1:
            valid_end = n_samples
        else:
            valid_end = train_end + fold_size

        train_idx = np.arange(0, train_end)
        valid_idx = np.arange(train_end, valid_end)

        if len(valid_idx) > 0:
            splits.append((train_idx, valid_idx))

        train_end = valid_end

    return splits


wf_splits = make_expanding_window_splits(
    n_samples=len(train_val_df),
    n_splits=N_WF_SPLITS,
    min_train_fraction=MIN_TRAIN_FRACTION
)

logger.info(f"Number of walk-forward folds: {len(wf_splits)}")


# ============================================================
# 12) BUILD MODEL
# ============================================================
def build_model(params):
    return XGBRegressor(
        n_estimators=int(params["n_estimators"]),
        max_depth=int(params["max_depth"]),
        learning_rate=float(params["learning_rate"]),
        subsample=float(params["subsample"]),
        colsample_bytree=float(params["colsample_bytree"]),
        min_child_weight=float(params["min_child_weight"]),
        reg_alpha=float(params["reg_alpha"]),
        reg_lambda=float(params["reg_lambda"]),
        gamma=float(params["gamma"]),
        objective="reg:squarederror",
        random_state=SEED,
        n_jobs=-1,
        early_stopping_rounds=EARLY_STOPPING_ROUNDS,
        tree_method="hist",
        eval_metric="rmse"
    )


# ============================================================
# 13) SINGLE TRIAL WALK-FORWARD EVALUATION
# ============================================================
def evaluate_single_trial_walk_forward(trial_name, params, X_all, y_all, date_series, splits):
    trial_log_path = os.path.join(LOG_DIR, f"{trial_name}.log")
    tlog = build_logger(trial_log_path, f"{trial_name}_logger")

    tlog.info("=" * 100)
    tlog.info(f"Single Trial: {trial_name}")
    tlog.info(f"Params: {params}")

    fold_rows = []

    for fold_id, (tr_idx, va_idx) in enumerate(splits, start=1):
        X_tr = X_all.iloc[tr_idx]
        y_tr = y_all.iloc[tr_idx]
        X_va = X_all.iloc[va_idx]
        y_va = y_all.iloc[va_idx]

        model = build_model(params)
        model.fit(
            X_tr, y_tr,
            eval_set=[(X_va, y_va)],
            verbose=False
        )

        va_pred = model.predict(X_va)
        va_metrics = regression_metrics(y_va, va_pred)
        va_score = composite_score(va_metrics)

        best_iter = getattr(model, "best_iteration", None)
        if best_iter is None:
            best_iter = getattr(model, "best_ntree_limit", None)

        fold_row = {
            "trial_name": trial_name,
            "fold": fold_id,
            "train_start": str(date_series.iloc[tr_idx].min()),
            "train_end": str(date_series.iloc[tr_idx].max()),
            "valid_start": str(date_series.iloc[va_idx].min()),
            "valid_end": str(date_series.iloc[va_idx].max()),
            "best_iteration": best_iter,
            "rmse": va_metrics["rmse"],
            "mae": va_metrics["mae"],
            "r2": va_metrics["r2"],
            "direction_acc": va_metrics["direction_acc"],
            "pred_std": va_metrics["pred_std"],
            "true_std": va_metrics["true_std"],
            "std_ratio": va_metrics["std_ratio"],
            "corr": va_metrics["corr"],
            "abs_mean_ratio": va_metrics["abs_mean_ratio"],
            "nonzero_rate_1e4": va_metrics["nonzero_rate_1e4"],
            "nonzero_rate_5e4": va_metrics["nonzero_rate_5e4"],
            "nonzero_rate_1e3": va_metrics["nonzero_rate_1e3"],
            "score": va_score
        }
        fold_rows.append(fold_row)

        tlog.info(
            f"Fold {fold_id} | "
            f"rmse={va_metrics['rmse']:.6f} | "
            f"mae={va_metrics['mae']:.6f} | "
            f"r2={va_metrics['r2']:.6f} | "
            f"dir={va_metrics['direction_acc']:.4f} | "
            f"std_ratio={va_metrics['std_ratio']:.4f} | "
            f"corr={va_metrics['corr']:.4f} | "
            f"score={va_score:.6f} | "
            f"best_iteration={best_iter}"
        )

    fold_df = pd.DataFrame(fold_rows)

    summary = {
        "trial_name": trial_name,
        **params,
        "mean_rmse": fold_df["rmse"].mean(),
        "mean_mae": fold_df["mae"].mean(),
        "mean_r2": fold_df["r2"].mean(),
        "mean_direction_acc": fold_df["direction_acc"].mean(),
        "mean_std_ratio": fold_df["std_ratio"].mean(),
        "mean_abs_mean_ratio": fold_df["abs_mean_ratio"].mean(),
        "mean_corr": fold_df["corr"].mean(),
        "mean_nonzero_rate_1e4": fold_df["nonzero_rate_1e4"].mean(),
        "mean_nonzero_rate_5e4": fold_df["nonzero_rate_5e4"].mean(),
        "mean_nonzero_rate_1e3": fold_df["nonzero_rate_1e3"].mean(),
        "mean_score": fold_df["score"].mean(),
        "log_path": trial_log_path
    }

    return summary, fold_df


trial_name = "single_best_trial"
trial_summary, fold_df = evaluate_single_trial_walk_forward(
    trial_name=trial_name,
    params=BEST_PARAMS,
    X_all=X_train_val,
    y_all=y_train_val,
    date_series=train_val_df[DATE_COL],
    splits=wf_splits
)

fold_csv = os.path.join(RESULT_DIR, "single_trial_fold_results.csv")
fold_df.to_csv(fold_csv, index=False)

summary_df = pd.DataFrame([trial_summary])
summary_csv = os.path.join(RESULT_DIR, "single_trial_summary.csv")
summary_df.to_csv(summary_csv, index=False)

print("\n=== SINGLE TRIAL WALK-FORWARD SUMMARY ===")
display(summary_df)

best_trial_plot_paths = plot_best_trial_folds(
    fold_df=fold_df,
    out_path_prefix=os.path.join(PLOT_DIR, "single_trial_validation"),
    trial_name=trial_name
)


# ============================================================
# 14) FINAL TRAIN ON TRAIN+VAL
# ============================================================
def fit_final_model(params, X_all, y_all, date_all, final_es_ratio=0.10):
    n_all = len(X_all)
    es_size = max(50, int(n_all * final_es_ratio))
    train_size = n_all - es_size

    X_fit = X_all.iloc[:train_size]
    y_fit = y_all.iloc[:train_size]
    X_es = X_all.iloc[train_size:]
    y_es = y_all.iloc[train_size:]

    logger.info(f"Final fit train size: {len(X_fit)}")
    logger.info(f"Final fit eval size : {len(X_es)}")

    model = build_model(params)
    model.fit(
        X_fit, y_fit,
        eval_set=[(X_fit, y_fit), (X_es, y_es)],
        verbose=False
    )

    evals_result = model.evals_result()
    return model, train_size, es_size, evals_result


best_model, final_train_size, final_es_size, final_evals_result = fit_final_model(
    params=BEST_PARAMS,
    X_all=X_train_val,
    y_all=y_train_val,
    date_all=train_val_df[DATE_COL],
    final_es_ratio=FINAL_ES_RATIO
)

best_iteration = getattr(best_model, "best_iteration", None)
logger.info(f"Final model best_iteration: {best_iteration}")

final_learning_curve_path = plot_final_learning_curve(
    evals_result=final_evals_result,
    out_path=os.path.join(PLOT_DIR, "final_learning_curve.png")
)


# ============================================================
# 15) TEST EVALUATION
# ============================================================
test_pred = best_model.predict(X_test)
test_metrics = regression_metrics(y_test, test_pred)
test_score = composite_score(test_metrics)

print("\n=== FINAL TEST RESULT ===")
print(f"Trial name          : {trial_name}")
print(f"Test RMSE           : {test_metrics['rmse']:.6f}")
print(f"Test MAE            : {test_metrics['mae']:.6f}")
print(f"Test R2             : {test_metrics['r2']:.6f}")
print(f"Test Direction Acc  : {test_metrics['direction_acc']:.4f}")
print(f"Pred std            : {test_metrics['pred_std']:.6f}")
print(f"True std            : {test_metrics['true_std']:.6f}")
print(f"Std ratio           : {test_metrics['std_ratio']:.4f}")
print(f"Abs mean ratio      : {test_metrics['abs_mean_ratio']:.4f}")
print(f"Corr                : {test_metrics['corr']:.4f}")
print(f"Nonzero rate >1e-4  : {test_metrics['nonzero_rate_1e4']:.4f}")
print(f"Nonzero rate >5e-4  : {test_metrics['nonzero_rate_5e4']:.4f}")
print(f"Nonzero rate >1e-3  : {test_metrics['nonzero_rate_1e3']:.4f}")
print(f"Composite score     : {test_score:.6f}")


# ============================================================
# 16) SAVE TEST PREDICTIONS
# ============================================================
pred_df = pd.DataFrame({
    DATE_COL: test_df[DATE_COL].values,
    "Actual_y_ret_t1": y_test.values,
    "Predicted_y_ret_t1": test_pred,
})
pred_df["Residual"] = pred_df["Actual_y_ret_t1"] - pred_df["Predicted_y_ret_t1"]
pred_df["ActualSign"] = np.sign(pred_df["Actual_y_ret_t1"])
pred_df["PredSign"] = np.sign(pred_df["Predicted_y_ret_t1"])
pred_df["Hit"] = (pred_df["ActualSign"] == pred_df["PredSign"]).astype(int)

pred_csv = os.path.join(RESULT_DIR, "test_predictions.csv")
pred_df.to_csv(pred_csv, index=False)

print("\n=== TEST PREDICTIONS (HEAD) ===")
display(pred_df.head(20))


# ============================================================
# 17) FEATURE IMPORTANCE
# ============================================================
importance_df = pd.DataFrame({
    "feature": FEATURE_COLS,
    "importance": best_model.feature_importances_
}).sort_values("importance", ascending=False).reset_index(drop=True)

importance_csv = os.path.join(RESULT_DIR, "feature_importance.csv")
importance_df.to_csv(importance_csv, index=False)

print("\n=== TOP 20 FEATURE IMPORTANCE ===")
display(importance_df.head(20))

fig = plt.figure(figsize=(10, 8))
top_imp = importance_df.head(20).iloc[::-1]
plt.barh(top_imp["feature"], top_imp["importance"])
plt.title("Top 20 Feature Importances")
plt.xlabel("Importance")
save_and_show(fig, os.path.join(PLOT_DIR, "feature_importance_top20.png"))


# ============================================================
# 18) PLOTS
# ============================================================
fig = plt.figure(figsize=(12, 5))
plt.plot(pred_df["Actual_y_ret_t1"].values[:200], label="Actual")
plt.plot(pred_df["Predicted_y_ret_t1"].values[:200], label="Predicted")
plt.title("Actual vs Predicted y_ret_t1 (first 200 test samples)")
plt.xlabel("Sample")
plt.ylabel("y_ret_t1")
plt.legend()
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "test_actual_vs_pred.png"))

fig = plt.figure(figsize=(6, 6))
plt.scatter(pred_df["Actual_y_ret_t1"], pred_df["Predicted_y_ret_t1"], alpha=0.5)
mn = min(pred_df["Actual_y_ret_t1"].min(), pred_df["Predicted_y_ret_t1"].min())
mx = max(pred_df["Actual_y_ret_t1"].max(), pred_df["Predicted_y_ret_t1"].max())
plt.plot([mn, mx], [mn, mx], linestyle="--")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Scatter: Actual vs Predicted")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "scatter_actual_vs_pred.png"))

fig = plt.figure(figsize=(12, 4))
plt.plot(pred_df["Residual"].values)
plt.axhline(0, linestyle="--")
plt.title("Residuals over Test Samples")
plt.xlabel("Sample")
plt.ylabel("Residual")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "residual_over_time.png"))

fig = plt.figure(figsize=(8, 4))
plt.hist(pred_df["Residual"].values, bins=40)
plt.title("Residual Histogram")
plt.xlabel("Residual")
plt.ylabel("Count")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "residual_hist.png"))

fig = plt.figure(figsize=(8, 4))
plt.hist(pred_df["Actual_y_ret_t1"].values, bins=40, alpha=0.6, label="Actual")
plt.hist(pred_df["Predicted_y_ret_t1"].values, bins=40, alpha=0.6, label="Predicted")
plt.title("Actual vs Predicted Distribution")
plt.xlabel("Return")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "actual_vs_pred_distribution.png"))

pred_df["CumHitRate"] = pred_df["Hit"].expanding().mean()
fig = plt.figure(figsize=(12, 4))
plt.plot(pred_df["CumHitRate"].values)
plt.axhline(0.5, linestyle="--")
plt.title("Cumulative Direction Accuracy")
plt.xlabel("Sample")
plt.ylabel("Accuracy")
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "cumulative_direction_accuracy.png"))

pred_df["RollingPredStd_30"] = pred_df["Predicted_y_ret_t1"].rolling(30).std()
pred_df["RollingTrueStd_30"] = pred_df["Actual_y_ret_t1"].rolling(30).std()
fig = plt.figure(figsize=(12, 4))
plt.plot(pred_df["RollingPredStd_30"].values, label="Pred rolling std (30)")
plt.plot(pred_df["RollingTrueStd_30"].values, label="True rolling std (30)")
plt.title("Rolling Std Comparison")
plt.xlabel("Sample")
plt.ylabel("Std")
plt.legend()
plt.grid(True)
save_and_show(fig, os.path.join(PLOT_DIR, "rolling_std_comparison.png"))


# ============================================================
# 19) SAVE FINAL SUMMARY
# ============================================================
final_summary = {
    "model_type": "XGBRegressor_single_trial",
    "trial_name": trial_name,
    "best_params": BEST_PARAMS,
    "num_features": len(FEATURE_COLS),
    "feature_cols": FEATURE_COLS,
    "n_walk_forward_splits": N_WF_SPLITS,
    "final_best_iteration": None if best_iteration is None else int(best_iteration),
    "final_train_size_before_es": int(final_train_size),
    "final_es_size": int(final_es_size),
    "walk_forward_mean_rmse": float(trial_summary["mean_rmse"]),
    "walk_forward_mean_mae": float(trial_summary["mean_mae"]),
    "walk_forward_mean_r2": float(trial_summary["mean_r2"]),
    "walk_forward_mean_direction_acc": float(trial_summary["mean_direction_acc"]),
    "walk_forward_mean_std_ratio": float(trial_summary["mean_std_ratio"]),
    "walk_forward_mean_corr": float(trial_summary["mean_corr"]),
    "walk_forward_mean_score": float(trial_summary["mean_score"]),
    "test_rmse": float(test_metrics["rmse"]),
    "test_mae": float(test_metrics["mae"]),
    "test_r2": float(test_metrics["r2"]),
    "test_direction_acc": float(test_metrics["direction_acc"]),
    "test_pred_std": float(test_metrics["pred_std"]),
    "test_true_std": float(test_metrics["true_std"]),
    "test_std_ratio": float(test_metrics["std_ratio"]),
    "test_abs_mean_ratio": float(test_metrics["abs_mean_ratio"]),
    "test_corr": float(test_metrics["corr"]),
    "test_nonzero_rate_1e4": float(test_metrics["nonzero_rate_1e4"]),
    "test_nonzero_rate_5e4": float(test_metrics["nonzero_rate_5e4"]),
    "test_nonzero_rate_1e3": float(test_metrics["nonzero_rate_1e3"]),
    "test_composite_score": float(test_score),
    "single_trial_summary_csv": summary_csv,
    "single_trial_fold_csv": fold_csv,
    "pred_csv": pred_csv,
    "feature_importance_csv": importance_csv,
    "main_log_path": main_log_path,
    "best_trial_validation_plots": best_trial_plot_paths,
    "final_learning_curve_plot": final_learning_curve_path
}

with open(os.path.join(RESULT_DIR, "final_summary.json"), "w", encoding="utf-8") as f:
    json.dump(final_summary, f, indent=2)

print("\n=== FINAL SUMMARY ===")
display(pd.DataFrame([final_summary]))

logger.info("All done.")
```
